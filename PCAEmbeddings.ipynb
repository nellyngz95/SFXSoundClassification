```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import os
import torch
import pandas as pd
```

Having all tfhe embeddings in csv for the class

```python
embed1 = pd.read_csv('/Users/nellygarcia/Documents/SFXSoundClassification/validation.csv')
embed2 = pd.read_csv('/Users/nellygarcia/Documents/SFXSoundClassification/train.csv')
embed3 = pd.read_csv('/Users/nellygarcia/Documents/SFXSoundClassification/test.csv')
EmbedFinal = pd.concat([embed1, embed2, embed3])
EmbedFinal = EmbedFinal.dropna()
# Directory containing the embeddings
embedding_dir = '/Users/nellygarcia/Documents/GitHub/PythonNelly/VGGSoundClassification/embeddings/'
#EmbedFinal.to_csv('/Users/nellygarcia/Documents/SFXSoundClassification/EmbedFinal.csv')
EmbedFinal
```

```python
embeddings = []
labels = [] 
for index,row in EmbedFinal.iterrows():
    filename = row['file']
    label = row['class']
     # Load each embedding file
    embedding_path = os.path.join(embedding_dir, filename)
    embedding = torch.load(embedding_path)
        
    # Append embedding and label
    embeddings.append(embedding)
    labels.append(label)

# Convert the lists to numpy arrays
embeddings = torch.stack(embeddings)
labels = np.array(labels)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(embeddings.numpy())

# Get unique labels
unique_labels = np.unique(labels)

# Plot the PCA results with different colors for each class
plt.figure(figsize=(10, 6))
for label in unique_labels:
    indices = np.where(labels == label)
    plt.scatter(pca_result[indices, 0], pca_result[indices, 1], label=label, alpha=0.5)

plt.title('PCA Visualization of Embeddings with Class Labels')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)

```
